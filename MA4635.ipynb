```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def yoink_csv(name: str):
  '''
  Gets the csv from the islr site given a name of a csv.
  options are here as long as they end in .csv
  https://www.statlearning.com/resources-python
  '''
  url = f'https://www.statlearning.com/s/{name}.csv'
  try:
    res = requests.get(url)
    if res.status_code == 200:
      with open (f'{name}.csv', 'wb') as f:
        f.write(res.content)
    else:
      print('conn failed')
  except Exception as e:
    print(e)
```
